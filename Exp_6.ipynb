```python
import shutil
import hashlib
import os

source = input("Enter uploaded file name: ")
destination = "copy_" + source

# Create a dummy file if it doesn't exist, for demonstration purposes
if not os.path.exists(source):
    print(f"File '{source}' not found. Creating a dummy file for demonstration.")
    with open(source, "w") as f:
        f.write("This is a dummy file content.\n")
        f.write("Another line of content.\n")

shutil.copy2(source, destination)

def get_hash(filename):
    sha = hashlib.sha256()

    with open(filename, "rb") as f:
        while True:
            data = f.read(4096)
            if not data:
                break
            sha.update(data)

    return sha.hexdigest()

original_hash = get_hash(source)
copy_hash = get_hash(destination)

print("Original Hash :", original_hash)
print("Copied Hash   :", copy_hash)

if original_hash == copy_hash:
    print("Integrity Verified ✓")
else:
    print("Integrity Failed ✗")
```
