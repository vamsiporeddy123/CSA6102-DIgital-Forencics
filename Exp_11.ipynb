```python
from collections import defaultdict

records = int(input("Enter number of network records: "))

ports = defaultdict(set)

for i in range(records):
    print("\nRecord", i + 1)

    ip = input("Source IP: ")
    port = int(input("Destination Port: "))

    ports[ip].add(port)

print("\nScanning Report")
print("----------------------")

found = False

for ip in ports:
    if len(ports[ip]) >= 5:
        print(ip, "is a Possible Port Scanner.")
        print("Ports Accessed:", sorted(ports[ip]))
        found = True

if not found:
    print("No port scanning behaviour detected.")
```
