```python
from collections import Counter

failed_attempts = []

n = int(input("Enter number of login records: "))

for i in range(n):
    print("\nRecord", i + 1)

    ip = input("IP Address: ")
    status = input("Status (SUCCESS/FAIL): ").upper()

    if status == "FAIL":
        failed_attempts.append(ip)

counter = Counter(failed_attempts)

print("\nSuspicious IP Addresses")
print("-------------------------")

attack_found = False

for ip, count in counter.items():
    if count >= 3:
        print(ip, "->", count, "failed login attempts")
        attack_found = True

if not attack_found:
    print("No brute-force attack detected.")
```
