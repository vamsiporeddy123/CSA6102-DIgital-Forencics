```python
import os
import time


file_name = "evidence.txt"

with open(file_name, "w") as file:
    file.write("Original Evidence")

print("File Created!")

time.sleep(3)

with open(file_name, "a") as file:
    file.write("\nModified by User")

print("File Modified!")

created_time = os.path.getctime(file_name)
modified_time = os.path.getmtime(file_name)

print("\n--- File Metadata ---")
print("Created Time :", time.ctime(created_time))
print("Modified Time:", time.ctime(modified_time))

if modified_time > created_time:
    print("\nLocard's Exchange Principle Demonstrated!")
    print("The file modification left a metadata trace.")
```
