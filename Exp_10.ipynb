```python
print("Paste the raw email header.")
print("Press Enter twice to finish.\n")

lines = []

while True:
    line = input()

    if line == "":
        break

    lines.append(line)

header = "\n".join(lines)

print("\nAnalysis Report")
print("------------------------")

if "Reply-To:" in header:
    print("Reply-To field detected.")

if "Received:" in header:
    print("Received path available.")

if "unknown" in header.lower():
    print("Unknown sender path detected.")

if "spf=fail" in header.lower():
    print("SPF verification failed.")

if "dkim=fail" in header.lower():
    print("DKIM verification failed.")

print("\nHeader analysis completed.")
```
