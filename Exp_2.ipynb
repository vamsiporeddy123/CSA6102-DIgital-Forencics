```python
import hashlib


with open("evidence.txt", "w") as f:
    f.write("Digital Evidence")

with open("evidence.txt", "rb") as f:
    file_hash = hashlib.sha256(f.read()).hexdigest()

print("SHA-256 Hash:")
print(file_hash)

print("\nChain of Custody Verified: File Integrity Maintained")
```
