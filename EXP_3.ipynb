```python
sender = input("Enter sender email: ")
subject = input("Enter subject: ")
body = input("Enter email body: ")

score = 0


if "unknown" in sender.lower() or "@fake" in sender.lower():
    score += 1

if "urgent" in subject.lower() or "verify" in subject.lower():
    score += 1

if "click here" in body.lower() or "password" in body.lower():
    score += 1

if score >= 2:
    print("\nPhishing Email Detected!")
else:
    print("\nEmail appears Safe.")
```
