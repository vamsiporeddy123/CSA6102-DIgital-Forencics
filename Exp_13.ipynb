```python
case_name = "Phishing Investigation"
investigator = "P C Vamsidhar Reddy"
date = "21-07-2026"

findings = [
    "Suspicious login attempts detected from IP 192.168.1.10.",
    "SHA-256 hash verified successfully.",
    "Spoofed email header identified.",
    "Port scanning activity detected.",
    "Recovered deleted PDF file from storage image."
]
report = f"""
=========================================
        DIGITAL FORENSIC REPORT
=========================================

Case Name      : {case_name}
Investigator   : {investigator}
Date           : {date}

Investigation Findings:
"""

for i, item in enumerate(findings, start=1):
    report += f"{i}. {item}\n"

report += """
Conclusion:
The digital evidence was successfully analyzed.
The identified security incidents have been documented.

Status: Investigation Completed
=========================================
"""
with open("Digital_Forensic_Report.txt", "w") as file:
    file.write(report)
print(report)

print("Report saved as 'Digital_Forensic_Report.txt'")
```
